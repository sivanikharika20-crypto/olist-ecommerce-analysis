## Handling missing values

```python
# Handling missing values in product category
products['product_category_name'] = products['product_category_name'].fillna('Unknown')
```

```python
products.isnull().sum()
```

```python
# Handle missing review comments
reviews['review_comment_message'].fillna("No Comment", inplace=True)
```

```python
reviews.isnull().sum()
```

## Removing duplicates

```python
customers.drop_duplicates(inplace=True)
orders.drop_duplicates(inplace=True)
order_items.drop_duplicates(inplace=True)
```

```python
customers.duplicated().sum()
```

## converting data type

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
```

```python
orders.info()
```

## Feature Engineering

```python
orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
orders['order_day'] = orders['order_purchase_timestamp'].dt.day_name()
```

## Checking outliers

## Data merging

```python
# Merge Orders and Customers
df = orders.merge(customers, on='customer_id', how='left')
```

```python
df.head()
```

```python
df.shape
```

```python
# merge order items
df = df.merge(order_items, on='order_id', how='left')
```

```python
df.head()
```

```python
# Merge Products Dataset
df = df.merge(products, on='product_id', how='left')
df
```

```python

# Merge Payments Dataset
df = df.merge(payments, on='order_id', how='left')
df
```

```python

# Merge Reviews Dataset
df = df.merge(reviews[['order_id','review_score']], on='order_id', how='left')
df
```

```python
# Merge Sellers Dataset
df = df.merge(sellers, on='seller_id', how='left')
df
```

```python
df.head
```

## EDA - Exploratory Data Analysis

```python
# Converting purchase timestamp to datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

# Creating new time features
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_day'] = df['order_purchase_timestamp'].dt.day_name()
```

```python
# Orders Distribution by Year

df['order_year'].value_counts().sort_index().plot(kind='bar')
plt.title("Number of Orders by Year")
plt.xlabel("Year")
plt.ylabel("Total Orders")
plt.show()
```

```python
# Orders Trend by Month

df['order_month'].value_counts().sort_index().plot(kind='line', marker='o')

plt.title("Monthly Order Trend")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Top 10 Most Purchased Product Categories

top_categories = df['product_category_name'].value_counts().head(10)

sns.barplot(x=top_categories.values, y=top_categories.index)

plt.title("Top 10 Product Categories by Orders")
plt.xlabel("Number of Orders")
plt.ylabel("Product Category")
plt.show()
```

```python
# Review Score Distribution

df['review_score'].value_counts().sort_index().plot(kind='bar')

plt.title("Customer Review Score Distribution")
plt.xlabel("Review Score")
plt.ylabel("Count")
plt.show()
```

```python
# Payment Methods Used by Customers

sns.countplot(data=df, x='payment_type')

plt.title("Distribution of Payment Methods")
plt.xlabel("Payment Type")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
# Distribution of Product Prices

sns.histplot(df['price'], bins=40)

plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Sellers Distribution by State

top_states = df['seller_state'].value_counts().head(10)

sns.barplot(x=top_states.index, y=top_states.values)

plt.title("Top Seller States")
plt.xlabel("State")
plt.ylabel("Number of Sellers")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Cities with Highest Number of Orders

top_cities = df['customer_city'].value_counts().head(10)

sns.barplot(x=top_cities.values, y=top_cities.index)

plt.title("Top 10 Cities by Number of Orders")
plt.xlabel("Number of Orders")
plt.ylabel("City")
plt.show()
```

```python
# Distribution of Freight Costs

sns.histplot(df['freight_value'], bins=30)

plt.title("Freight Cost Distribution")
plt.xlabel("Freight Cost")
plt.ylabel("Frequency")
plt.show()
```

```python
# Orders by Day of Week

sns.countplot(data=df, x='order_day')

plt.title("Orders by Day of Week")
plt.xlabel("Day")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
# Revenue Trend by Month

revenue_by_month = df.groupby('order_month')['payment_value'].sum()

revenue_by_month.plot(kind='line', marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Top Product Categories by Revenue

top_category_revenue = df.groupby('product_category_name')['payment_value'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_category_revenue.values, y=top_category_revenue.index)

plt.title("Top Product Categories by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product Category")
plt.show()
```

```python
# Top Sellers by Revenue

top_sellers = df.groupby('seller_id')['payment_value'].sum().sort_values(ascending=False).head(10)

top_sellers.plot(kind='bar')

plt.title("Top 10 Sellers by Revenue")
plt.xlabel("Seller ID")
plt.ylabel("Total Revenue")
plt.show()
```

```python
# Order Status Distribution

sns.countplot(data=df, x='order_status')

plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()
```

3. Top Product Categories

Certain product categories dominate order volume, suggesting higher customer demand in those segments.

4. Customer Satisfaction

Most review scores are 4 and 5, indicating generally high customer satisfaction.

5. Delivery Performance

Orders with longer delivery times tend to have lower review scores, suggesting delivery speed affects customer satisfaction.

6. Revenue Distribution

A few cities generate a large portion of total revenue, indicating key geographic markets.

7. Seller Performance

Some sellers contribute significantly to total sales, highlighting top-performing vendors.
